# **What is a Neural Network?**

A neural network is a computation model inspired by the human brain. It consists of interconnected nodes, called neurons, organized into layers. These connections have weights that can be adjusted to improve the network's performance in learning and predicting outcomes.

**Prerequisites**
* NumPy: For numerical computing
* TensorFlow: For creating the neural network
* Keras: A high-level neural networks API

# **Import the required libraries**

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

# **Load and Prepare the Data**
In this tutorial, we will use the popular MNIST dataset containing 70,000 grayscale images of handwritten digits. The dataset is divided into 60,000 training images and 10,000 testing images. The goal is to train our neural network to recognize these handwritten digits.

Load the data using the following code:

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

Next, normalize the image data by dividing each pixel value by 255. This scales the pixel values between 0 and 1, which helps the neural network converge faster during training.

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

# **Create the Neural Network Model**
Create a neural network model using the Keras API. In this example, we will create a simple feedforward neural network with one input layer, hidden layer, and output layer.

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

# **Compile the Model**
Compile the model by specifying the optimizer, loss function, and evaluation metric:

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

# **Train the Model**
Train the neural network using the fit() method. Specify the training data, the number of epochs, and the batch size.

```python
history = model.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python
print(history.history.keys())
```

# **Evaluate the Model**
After training, evaluate the model's performance on the test dataset:

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print('Test accuracy:', test_accuracy)
```
